Antes de reemplazar los valores ausentes de la columna `churn_date` vamos a generar una nueva columna que contenga las categorías active para aquellos clientes que todavía tienen un plan e inactive para aquellos que cancelaron el plan. Para esto crearemos una función `churn_categories` que clasificará los valores NaT como clientes activos y las fechas en las que termina de usar el servicio como clientes inactivos.

```python
def churn_categories(churn_date):
    if pd.isna(churn_date):
        return 'active'
    return 'inactive'
```

### Ingresos mensuales por usuario

En base a los minutos utilizados al mes, la cantidad de SMS enviados al mes y el volumen de datos utilizado al mes, tenemos que calcular los ingresos mensuales que genera cada usuario de Megaline. Para esto vamos a revisar nuevamente nuestra tabla plans y estableceremos las recargas que cada plan presenta cuando los usuarios sobrepasan el límite del paquete gratuito.

```python
plans
```

Podemos observar que el plan `surf` tiene un costo mensual de 20 dólares, si se sobrepasan los 500 minutos, los 15 GB y los 50 mensajes que incluye el plan, las recargas serán las siguientes:

- 1 minuto: 3 centavos
- 1 SMS: 3 centavos
- 1 GB: 10 dólares

En el caso del plan `ultimate`, el costo mensual es de 70 dólares, si se excede el límite de 3000 minutos, 30 GB y 1000 mensajes, las recargas al valor mensual serán las que se presentan a continuación:

- 1 minuto: 1 centavo
- 1 SMS: 1 centavo
- 1 GB: 7 dólares

Para poder establecer el ingreso mensual por usuario vamos a crear una función `monthly_income_func`, la función tomará los valores de las columnas `plan`, `min_per_month`, `messages_per_month` y `gb_per_month` y calculará el ingreso de cada usuario. Para esto se establecerán tres contadores en cero a los cuales se les irá añadiendo los valores cuando el cliente excede los límites del paquete. Los contadores serán `min_total`, `mess_total` y `gb_total`, a cada contador se le añadirán los excedentes establecidos para minutos, mensajes y datos.

El cálculo del excedente se realizará restando el límite del paquete del número total de llamadas, mensajes y datos, este resultado se multiplicará por el valor de tarifa establecido por Megaline, que mencionamos anteriomente. Finalmente se sumarán los valores obtenidos por los tres contadores al cargo mensual según la tarifa de llamadas, en este caso será 20 para surf y 70 para ultimate, este resultado se guardará en la variable income, la cual nos dará el valor de ingreso mensual que estamos buscando.

```python
def monthly_income_func(row):
    min_total = 0
    mess_total = 0
    gb_total = 0

    plan =  row['plan']
    min_per_month = row['min_per_month']
    messages_per_month =  row['messages_per_month']
    gb_per_month = row['gb_per_month']

    if plan == 'surf':

        if min_per_month > 500:
            min_total += (min_per_month - 500) * 0.03

        if messages_per_month > 50:
            mess_total += (messages_per_month -50) * 0.03

        if gb_per_month > 15:
            gb_total +=  (gb_per_month - 15) * 10


        income = 20 + min_total + mess_total + gb_total


    if plan == 'ultimate':

        if min_per_month > 3000:
            min_total += (min_per_month - 3000) * 0.01

        if messages_per_month > 1000:
            mess_total += (messages_per_month - 1000) * 0.01

        if gb_per_month > 30:
            gb_total += (gb_per_month - 30) * 7


        income = 70 + min_total + mess_total + gb_total

    return income
```

Probamos que la función esté trabajando correctamente, para esto la aplicamos a la fila 3 de nuestro dataset `users_final`.

```python
monthly_income_func(users_final.iloc[3])
```

Ahora creamos una nueva columna que guarde los resultados de ingresos mensuales por usuario `monthly_income` y usamos el método apply con axis=1 para aplicar nuestra función `monthly_income_func` a nuestro dataset users_final.

```python
users_final['monthly_income'] = users_final.apply(monthly_income_func, axis=1)
```

Comprobamos que la nueva columna se haya generado correctamente, llamando al método head en nuestro dataset users_final.

```python
def region_id(city):

    if 'NY-NJ' in city or 'NY' in city or 'NJ' in city:
        return 'ny-nj'
    return 'other'
```

Verificamos que la función esté trabajando correctamente para eso aplicamos la función a una de los usuarios de estos estados.

```python
users_final['region'] = users_final['city'].apply(region_id)
```

A partir de cada plan, nos devolverá un histograma con la función hist de pyplot, un gráfico de densidad con la función kdeplot de seaborn, y los valores de media, mediana, varianza y desviación estándar de las columnas `min_per_month`, `calls_per_month`, `messages_per_month` y `gb_per_month`. La función contará con los siguientes parámetros:

### Estudio de valores atípicos

Al analizar la varianza y desviación estándar de las columnas `min_per_month`, `messages_per_month` y `gb_per_month`, se registran valores altos de dispersión para algunas de estas columnas, así que vamos a analizar más a fondo el comportamiento de valores atípicos para tomar una decisión sobre conservarlos o eliminarlos. Construiremos diagramas de caja que nos permitan identificar el rango intercuartílico y establecer los límites de nuestro dataset sin valores atípicos.

Crearemos la función `boxplot_function` que nos devolverá un diagrama de cajas con los límites del rango intercuartílico superior e inferior y además nos presentará las descripciones numéricas del método describe.

```python
def boxplot_function(df, column, ylim=(0,0)):
    Q1= df[column].quantile(0.25)
    Q3= df[column].quantile(0.75)
    IQR = Q3 - Q1

    df.boxplot(column)
    plt.hlines(y=[(Q1-1.5*IQR),(Q3+1.5*IQR)], xmin=0.9, xmax=1.1, color='red')
    plt.ylim(ylim)
    plt.show()

    print('Límite superior:', Q3+1.5*IQR)
    print('Límite inferior:', Q1-1.5*IQR)
    print()
    print(df[column].describe())
```

Vamos a empezar estudiando los valores atípicos de los clientes surf, para esto aplicaremos la función `boxplot_function` a las columnas `min_per_month`, `messages_per_month` y `gb_per_month` del dataslice `surf_users`

```python
boxplot_function(surf_users, 'min_per_month', (-200,1200))
```

```python
boxplot_function(surf_users, 'messages_per_month', (-70, 160))
```

```python
boxplot_function(surf_users, 'gb_per_month', (-5, 50))
```

Para la columna `min_per_month` del dataset `surf_users`, el límite superior del rango intercuartílico se presenta en 1036 minutos. En el caso de la columna `messages_per_month`, se puede observar el sesgo hacia la derecha de los datos con un límite superior del IQR en 113. Para el caso se la columna `gb_per_month`, se observan valores atípicos por tanto por encima de 34.5.

## Prueba de hipótesis

### Hipótesis 1: El ingreso promedio de los usuarios de las tarifas Ultimate y Surf difiere

Nuestra primera hipótesis establece que el ingreso promedio de los usuarios de las tarifas Ultimate y Surf difiere. Primero revisemos los valores promedio de ingresos para los usuarios de surf y ultimate. Para esto generaremos nuevos dataslices a partir de nuestro dataframe `users_filtered` sin valores atípicos.

```python
surf_filtered = users_filtered.query('plan == "surf"')
surf_filtered.info()
```

```python
ultimate_filtered = users_filtered.query('plan == "ultimate"')
ultimate_filtered.info()
```

```python
print('Ingreso promedio mensual plan Surf:', surf_filtered['monthly_income'].mean())
```

```python
print('Ingreso promedio mensual plan Ultimate:', ultimate_filtered['monthly_income'].mean())
```

En base a estos valores podríamos establecer que los usuarios de ultimate generan más ingresos que los usuarios de surf, pero no podemos afirmar esto sin realizar primeramente una prueba de hipótesis que analizará más a fondo las varianzas de las muestras y establecerá si estas diferencias son significativas. Es así que vamos a establecer nuestra hipótesis nula (H0) y nuestra hipótesis alternativa de la siguiente manera:

- H0 = *El ingreso promedio entre los usuarios de los planes Surf y Ultimate es igual*
- H1 = *El ingreso promedio entre los usuarios de los planes Surf y Ultimate difiere*

La hipótesis nula siempre va a indicar que no hay diferencias entre las variables, mientras que la hipótesis alternativa indicará que existen cambios o diferencias. Antes de aplicar la prueba de hipótesis vamos a establecer si las varianzas de las muestras son iguales o diferentes, así que llamaremos a np.var de la librería numpy y calcularemos las varianzas del ingreso mensual para los clientes surf y ultimate.

```python
print('Varianza ingreso mensual plan Surf:', np.var(surf_filtered['monthly_income']))
```

```python
print('Varianza ingreso mensual plan Ultimate:', np.var(ultimate_filtered['monthly_income']))
```

Parecería que existen diferencias entre las varianzas de los planes surf y ultimate. Para comprobarlo utilizaremos el test de igualdad de varianzas de Levene, una prueba estadística inferencial que evalúa la homocedasticidad o igualdad de varianzas, estableciendo como hipótesis nula la homogeneidad de varianzas entre dos muestras. Vamos a considerar un nivel de significancia de 0.05, si el valor p resultante es mayor a 0.05 entonces aceptaremos la hipótesis nula y si este valor es inferior a 0.05 aceptaremos la hipótesis alternativa y las varianzas no serán iguales.

Aplicamos entonces el test de Levene para nuestros dos grupos.

```python
levene(surf_filtered['monthly_income'], ultimate_filtered['monthly_income'])
```

En base al análisis estadístico, se encontró que el valor p es inferior a 0.05, lo que nos lleva a rechazar la hipótesis nula y concluir que las varianzas para el ingreso promedio mensual difieren entre los clientes con el plan Surf y Ultimate.

Dado que se ha verificado la no igualdad de varianzas, procederemos a realizar nuestra prueba de hipótesis utilizando la prueba T-Student. Esta prueba nos permitirá determinar si las medias aritméticas de las dos poblaciones son iguales o no. Compararemos el valor p obtenido con un nivel de significancia (alpha) establecido en 0.05. Si el valor p de la prueba T es menor a alpha, rechazaremos la hipótesis nula y aceptaremos la hipótesis alternativa. Por el contrario, si el valor p es mayor a alpha, aceptaremos la hipótesis nula.

Para realizar este análisis, utilizaremos la función st.ttest_ind de la librería scipy.stats. Esta función tomará como argumentos las dos muestras que se están comparando: sample_1 representará el ingreso mensual de los usuarios del plan Surf, mientras que sample_2 representará el ingreso mensual de los usuarios del plan Ultimate. Además, estableceremos el argumento equal_var en False, ya que el test de Levene demostró diferencias en las varianzas de las poblaciones. Posteriormente, utilizaremos una estructura condicional if/else para determinar las condiciones de rechazo o aceptación de la hipótesis nula, en función del valor alpha establecido.

```python
sample_1 = surf_filtered['monthly_income']
sample_2 = ultimate_filtered['monthly_income']
alpha = 0.05

results = st.ttest_ind(sample_1, sample_2, equal_var=False)

print('pvalue:', results.pvalue)

if results.pvalue < alpha:
    print("Rechazamos la hipótesis nula")
else:
    print("No rechazamos la hipótesis nula")
```

```python
ttest,pvalue = st.mannwhitneyu(sample_1,sample_2, alternative="two-sided")
print("p-value:%.4f" % pvalue)
if pvalue <0.05:
    print("Rechazamos la hipótesis nula")
else:
    print("No rechazamos la hipótesis nula")
```

El valor p es inferior al nivel de significancia de 0.05 así que se rechaza la hipótesis nula y se establece que el ingreso promedio mensual difiere significativamente entre los usuarios de surf y ultimate, siendo ultimate el plan que más ingresos genera a la empresa Megaline. Hay que mencionar la baja varianza registrada para los usuarios de ultimate, lo que podría ser un indicador que son pocos los clientes que sobrepasan el límite de mensajes, minutos o datos, por lo que solo algunos realizan un pago superior a los 70 dólares del plan ya establecido. A su vez, los usuarios del plan surf tienden a sobrepasar los límite de consumo del plan, lo que sería la causa de la varianza tan alta registrada.

Al parecer los usuarios de otras regiones generan más ingresos mensuales que los usuarios de NY-NJ, pero no podemos afirmar esto sin realizar una prueba de hipótesis que establezca si estas diferencias son significativas. Vamos a establecer nuestras dos hipótesis de la siguiente manera:

- H0 = *El ingreso promedio entre los usuarios de Nueva York-Nueva Jersey y otras regiones es igual*
- H1 = *El ingreso promedio entre los usuarios de Nueva York-Nueva Jersey y otras regiones difiere*

Como ya lo mencionamos anteriormente, la hipótesis nula siempre indica la igualdad de las variables mientras que la hipótesis alternativa se relaciona con posibilidades diferentes. Una vez establecidas nuestras hipótesis, procedemos a realizar un análisis de igualdad de varianzas, que nos servirá para establecer diferencias entre las medias poblacionales posteriormente. Empezamos calculando la varianza para los clientes de Nueva York-Nueva Jersey y los de otras regiones, para lo cual utilizaremos el método np.var de la librería Numpy.

```python
print('Varianza ingreso mensual región NY-NJ:', np.var(ny_nj_users['monthly_income']))
```

```python
print('Varianza ingreso mensual otras regiones:', np.var(other_region_users['monthly_income']))
```

Para determinar si existe una diferencia significativa en las varianzas entre los usuarios de otras regiones y los usuarios de la región NY-NJ, se realizará el test de homogeneidad de varianzas de Levene. Utilizando un nivel de significancia o alpha de 0.05, se establecerá si el valor de p obtenido es mayor o menor a alpha. Si el valor de p es mayor a alpha, se aceptará la hipótesis nula, indicando que las varianzas son iguales. Por otro lado, si el valor de p es menor a alpha, se rechazará la hipótesis nula, sugiriendo que existe una diferencia significativa en las varianzas.

Para llevar a cabo esta prueba, se utilizará la función Levene de la librería scipy.stats, y se pasarán como argumentos las dos muestras a comparar.

```python
levene(ny_nj_users['monthly_income'], other_region_users['monthly_income'])
```

El valor p del test de Levene es inferior a alpha, por lo que las muestras no presentan igualdad de varianzas. Considerando esto realizaremos nuestra prueba de hipótesis para el ingreso promedio mensual por regiones, utilizaremos una prueba t-student para comparar las medias de las dos muestras y estableceremos nuestro nivel de significancia en 0.05. Si el valor p de la prueba t es superior a alpha aceptaremos la hipótesis nula y si es inferior rechazaremos la hipótesis nula y el ingreso promedio mensual difiere entre ambas muestras.

Llamamos entonces a la función estadística st.ttest_ind de la librería scipy.stats y pasaremos como argumentos las dos muestras: sample_1 serán los ingresos mensuales de usuarios de NY-NJ y sample_2 los ingresos de clientes de otras regiones. En este caso, el test de Levene demostró varianzas no iguales, por lo que el argumento equal_var se establecerá en falso o False. Para aceptar o rechazar nuestra hipótesis construiremos un bloque if/else.

```python
sample_1 = ny_nj_users['monthly_income']
sample_2 = other_region_users['monthly_income']
alpha = 0.05

results = st.ttest_ind(sample_1, sample_2, equal_var=False)

print('pvalue:', results.pvalue)

if results.pvalue < alpha:
    print("Rechazamos la hipótesis nula")
else:
    print("No rechazamos la hipótesis nula")
```

```python
ttest,pvalue = st.mannwhitneyu(sample_1,sample_2, alternative="two-sided")
print("p-value:%.4f" % pvalue)
if pvalue <0.05:
    print("Rechazamos la hipótesis nula")
else:
    print("No rechazamos la hipótesis nula")
```

Basado en el análisis estadístico realizado, se determinó que el valor p obtenido es superior al nivel de significancia establecido de 0.05. Por lo tanto, no se puede rechazar la hipótesis nula. En consecuencia, se concluye que no existen diferencias significativas en el ingreso promedio mensual entre los clientes de Nueva York-Nueva Jersey y los clientes de otras regiones. En otras palabras, se encontró evidencia de que los ingresos entre ambas regiones son similares.
